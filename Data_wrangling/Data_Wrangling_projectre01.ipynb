```python
import pandas as pd
```

```python
filename = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv(filename, names = headers)
df.head()
```

```python
import numpy as np

df.replace("?",np.nan,inplace = True)
```

```python
df.head()
```

```python
missing_data = df.isnull()
missing_data.head()
```

```python
missing_data.columns.values.tolist()
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

```python
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
```

```python
 df["normalized-losses"].replace(np.nan,avg_norm_loss, inplace = True)
```

```python
avg_bore = df["bore"].astype('float').mean(axis=0)

print(f"Average of bore: {avg_bore}")
```

```python
df["bore"].replace(np.nan,avg_bore,inplace = True)
```

```python
df["bore"].isnull().value_counts()
```

```python
avg_bore=df["stroke"].astype('float').mean(axis=0)
```

```python
df["stroke"].replace(np.nan,avg_bore,inplace = True)
```

```python
df["stroke"].isnull().value_counts()
```

```python
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
avg_horsepower
```

```python
df['horsepower'].replace(np.nan,avg_horsepower,inplace = True)
```

```python
df['horsepower'].isnull().value_counts()
```

```python
avg_peak_rpm=df['peak-rpm'].astype('float').mean()
```

```python
df['peak-rpm'].replace(np.nan,avg_peak_rpm,inplace=True)
```

```python
df['peak-rpm'].isnull().value_counts()
```

```python
df['num-of-doors'].value_counts()
```

```python
df['num-of-doors'].value_counts().idxmax()
```

```python
df['num-of-doors'].replace(np.nan,"four",inplace=True)
```

```python
df['num-of-doors'].isnull().any()
```

```python
df.dropna(subset=["price"],axis=0,inplace=True)

df.reset_index(drop=True,inplace=True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
df.dtypes
```

```python
df.head()
```

```python
# Convert mpg to L/100km by mathematical operation (235 divided by mpg)
df['city-L/100km'] = 235/df["city-mpg"]

# check your transformed data 
df.head()
```

```python
df["highway-mpg"]=235/df["highway-mpg"]

df.rename(columns={'highway-mpg':'highway-L/100km'},inplace=True)

df.head()
```

```python
df[["highway-L/100km"]]
```

```python
df['length']=df['length']/df['length'].max()
df['width']=df['width']/df['width'].max()
```

```python
df['height']=df['height']/df['height'].max()
```

```python
df[["length","width","height"]].head()
```

```python
df["horsepower"]=df["horsepower"].astype(int,copy=True)
```

```python
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df['horsepower'])

plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
bins=np.linspace(min(df['horsepower']),max(df['horsepower']),4)
bins
```

```python
group_names = ['Low','Medium','High']
```

```python
df['horsepower-binned'] = pd.cut(df['horsepower'],bins,labels=group_names,include_lowest=True)
df[['horsepower','horsepower-binned']].head(10)
```

```python

import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```
